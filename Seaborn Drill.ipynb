```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic = sns.load_dataset("titanic")
sns.set(style="darkgrid", palette="deep")
```

### Drill 1 - One Variable, Four Different Ways

```python
# box plot

plt.figure(figsize=(6,10))
box = sns.boxplot(y='fare',data=titanic)
plt.show()
```

```python
# histogram

hist = sns.distplot(titanic['fare'])
plt.show()
```

```python
violin = sns.violinplot(y="fare", data=titanic)
plt.show()
```

```python
point = sns.pointplot(y='fare',data=titanic)
plt.show()
```

### 2 continuous variables, plotted 3 different ways

```python
scatter = sns.lmplot(x='age',y='fare',data=titanic,fit_reg=False)
plt.show()
```

```python
kde = sns.jointplot(x='age',y=titanic[titanic['fare'] <= 60]['fare'],data=titanic,kind="kde") #subsetting the fares for a smoother histogram than you'd get by limiting y axis
plt.ylim(0,60)
plt.show()
```

```python
joint = sns.jointplot(x='age',y='fare',data=titanic,alpha=0.3)
plt.show()
```

### A continuous variable and categorical variable, 6 ways

```python
bar = sns.factorplot(x='embark_town',y='fare',data = titanic, kind='bar',ci=95)
plt.show()
```

```python
point = sns.pointplot(x='embark_town',y='fare',data=titanic, join=False, dodge=False)
```

```python
sns.stripplot(x='embark_town',y='fare',data=titanic)
plt.show()
```

```python
sns.boxplot(x='embark_town',y='fare',data=titanic)
plt.show()
```

```python
sns.violinplot(x='embark_town',y='fare',data=titanic)
plt.show()
```

```python
southampton = titanic[titanic['embark_town'] == 'Southampton']['fare']
cherbourg = titanic[titanic['embark_town'] == 'Cherbourg']['fare']
queenstown = titanic[titanic['embark_town'] == 'Queenstown']['fare']

towndist = sns.distplot(southampton)
plt.title('Southampton')
plt.xlim(0,300)
plt.subplots_adjust(wspace=0.1)

plt.show()

towndist = sns.distplot(cherbourg)
plt.title('Cherbourg')
plt.xlim(0,300)
plt.subplots_adjust(top=0.9)

plt.show()

towndist = sns.distplot(queenstown)
plt.title('Queenstown')
plt.xlim(0,300)
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
facet = sns.FacetGrid(titanic,row='embark_town')
facet.map(plt.hist,"fare")
plt.subplots_adjust(left=0.125,right=2)
```

```python
facet = sns.FacetGrid(titanic,col='embark_town')
facet.map(plt.hist,"fare")
plt.subplots_adjust(left=0.125,right=0.9)
```

```python
dist = sns.distplot(titanic[titanic['embark_town'] == 'Southampton']['fare'],hist=False)
dist = sns.distplot(titanic[titanic['embark_town'] == 'Cherbourg']['fare'],hist=False)
dist = sns.distplot(titanic[titanic['embark_town'] == 'Queenstown']['fare'],hist=False)
```
