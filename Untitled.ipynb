We'll be parsing data related to Radiohead from the [iTunes public API](https://affiliate.itunes.apple.com/resources/documentation/itunes-store-web-service-search-api/) using scrapy.

```python
import scrapy
import json
from scrapy.crawler import CrawlerProcess


class iTunes(scrapy.Spider):
    name = "itunes"
    
    # Here is where we insert our API call.
    start_urls = [
        'https://itunes.apple.com/search?term=radiohead&entity=musicTrack'
        ]
    
    # Parsing the information we want from the json
    def parse(self, response):
        jsonresponse = json.loads(response.body_as_unicode())
        print(jsonresponse)
        for item in jsonresponse['results']:
            yield {
                'artistName':item['artistName']
                ,'trackName':item['trackName']
                ,'collectionPrice':item['collectionPrice']
                ,'trackPrice':item['trackPrice']
                ,'collectionName':item['collectionName']
                ,'trackTimeMillis':item['trackTimeMillis']}
                #,'trackExplicitness':item['trackExplicitness']
                #,'currency':item['currency']
                #,'contentAdvisoryRating':item['contentAdvisoryRating']}
               
process = CrawlerProcess({
    'FEED_FORMAT': 'json',
    'FEED_URI': 'itunes.json',
    'ROBOTSTXT_OBEY': False,
    'USER_AGENT': 'ThinkfulDataScienceBootcampCrawler (thinkful.com)',
    'AUTOTHROTTLE_ENABLED': True,
    'HTTPCACHE_ENABLED': True,
    'LOG_ENABLED': False,
})
                                         

# Starting the crawler with our spider.
process.crawl(iTunes)
process.start()
print('Done!')
```

```python
# Let's get this into a dataframe and do some quick visualizations
import pandas as pd

with open('itunes.json', 'r') as content_file:
    content = content_file.read()
records = json.loads(content)

radiohead = pd.DataFrame(records)
```

```python
import numpy as np
radiohead.pivot_table(radiohead,index='collectionName')['collectionPrice'].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Are all the albums priced the same?
#plt.hist(radiohead.pivot_table(radiohead,index='collectionName')['collectionPrice'].values)
#plt.xlabel(radiohead.pivot_table(radiohead,index='collectionName')['collectionPrice'].keys)
plt.figure(figsize=(20,10))
plt.title("Cost by Album")
x = pd.DataFrame(radiohead.pivot_table(radiohead,index='collectionName')['collectionPrice']).index
y = radiohead.pivot_table(radiohead,index='collectionName')['collectionPrice'].values
sns.barplot(x,y)
plt.show()
```

```python
rounded_song_length = np.round((radiohead['trackTimeMillis']/60000)*2)/2
plt.title("Radiohead Songs by Length")
plt.hist(rounded_song_length)
plt.show()
```
